```python
sen_tk=sent_tokenize(text)
print("senetnce tokenize the text \n")
print(sen_tk)
```

```python
word_tk=word_tokenize(text)
print("word tokenize the text \n")
print(word_tk)
```

```python
# remove stop words 

stop_words=set(stopwords.words("english"))
print("These are stop words in english: \n")
print(stop_words)
```

```python
filtered_words= [w for w in word_tk if not w in stop_words ]
print("The text after removing stop words: \n")
print(filtered_words)
```

```python
# stem 

port_stem=PorterStemmer()
Stemmed_words=[]

for w in filtered_words:
    Stemmed_words.append(port_stem.stem(w))
#print("Filtered Sentence: \n", filtered_words,"\n")
print("Stemmed Sentence: \n", Stemmed_words,"\n")


    
```

```python
#lemmatize
lem= WordNetLemmatizer()

lemm_words=[]
for i in range(len(filtered_words)):
    lemm_words.append(lem.lemmatize(filtered_words[i]))

print(lemm_words) 
    
```

```python
#nltk.download('averaged_perceptron_tagger')

post_taged_words = pos_tag(word_tk)
print(post_taged_words)
```
